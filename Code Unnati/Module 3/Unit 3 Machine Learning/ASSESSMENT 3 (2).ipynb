3. Create Python Tkinter GUI program to Predict house prices (using Linear regression)

```python
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

link = "https://github.com/Shreyansh81/House-Price-Prediction/raw/main/Housing%20Information.csv"
data = pd.read_csv(link)
df = pd.DataFrame(data)

numberOfNullsInDataframe = df.isnull().sum().sum()
print(f"Number of Null elements in DataFrame : {numberOfNullsInDataframe}")

df.corr() # Correlation between the features #
# We will only use features with maximum Correlation with the Price . #

features = ["Avg. Area Income","Avg. Area House Age","Avg. Area Number of Rooms","Area Population"]

y = np.array(df["Price"]).reshape(-1,1)
X = np.array(df[features])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

regression = LinearRegression()
regression.fit(X_train,y_train)

accuracy = regression.score(X_test,y_test)*100

print( f"The Accuracy of Model is : {accuracy:.2f}")

# As It's an Multiple Linear Regression Model the Graph of Model will have n-dimensions ,
# where n = number os features used to train a model .
# Hence it's almost impossible to plot a graph of n-dimension on a 2-D screen .
avgAreaIncome = float(input("Enter Average Area Income of your Area : "))
avgAreaHouseAge = float(input("Enter Average Area House Age of your Area : "))
avgAreaNumberOfRooms = float(input("Enter Average Area Number of Rooms of your Area : "))
areaPopulation = float(input("Enter Total Population of your Area : "))

xfeatures = [ avgAreaIncome , avgAreaHouseAge , avgAreaNumberOfRooms , areaPopulation ]
x_features = np.array(xfeatures).reshape(1,-1)

y_pred = str(regression.predict(x_features)[0][0])
print("The House Price in your area will be : " + y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
house_pred=pd.read_csv('housing.csv')
house_pred.head()
```

```python
X = house_pred[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms',
                'Avg. Area Number of Bedrooms','Area Population']]
y=house_pred.Price
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=101)
```

```python
yhat=lr.predict(X_test)
```

3.  Create Python Jupyter notebook of Sk-Learn program using Support Vector Machine(SVM) to classify cancer samples diagnosed as Benign or Malignant 

```python
from sklearn import svm
```

```python
df=pd.read_csv('data.txt')
df
```

```python
#df=df.dropna()
#X=df.drop("diagnosis",axis='columns')
X=df[['id', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
y=df.diagnosis
X_train, X_test, y_train, y_test = train_test_split(X , y , test_size=0.3, random_state=101)
```

```python
yhat=clf.predict(X_test)
```
